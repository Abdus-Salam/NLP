```python
import os
import pandas as pd 
from matplotlib import pyplot as plt
```

```python
data_dir = '../dataset'
#print(os.listdir(data_dir))
filename = os.path.join(data_dir, 'jena_climate_2009_2016.csv')
```

```python
nRowsRead = 1000 # specify 'None' if want to read whole file
# jena_climate_2009_2016.csv has 420551 rows in reality, but we are only loading/previewing the first 1000 rows
df = pd.read_csv(filepath_or_buffer=filename, delimiter=',', nrows = nRowsRead)
#df1.dataframeName = 'jena_climate_2009_2016.csv'
nRow, nCol = df.shape

print("There are %d rows and %d columns" %(nRow, nCol))
df.head(5)
```

# Plot correlation matrix usinmg matplotlib

```python
# create a figure
f = plt.figure(figsize=(19, 15))

# A = matrix = correlation matrix
# f.number = number of figure is 1
# output = color represent using value of the matrix = celcius * celcius = যে মান অাসবে সে অনুযায়ী color show করবে
matrix = df.corr()
#matrix.style.background_gradient(cmap='coolwarm')
plt.matshow(A = matrix, alpha=0.9, fignum=f.number)

# find out the x label and y label text
tricks_length = df.shape[1] # number of columns
column_names = df.columns # column name

plt.xticks(range(tricks_length), labels=column_names, fontsize=14, rotation=45)
plt.yticks(range(tricks_length), labels=column_names, fontsize=14)


cb = plt.colorbar()
cb.ax.tick_params(labelsize=14)
plt.title('Correlation Matrix', fontsize=16);
```

# Scatter matrix using matplotlib

```python
from pandas.tools.plotting import scatter_matrix
print(df.columns)
scatter_matrix(df.loc[:, 'p (mbar)' : 'Tpot (K)'], figsize=(12, 10), diagonal='kde')
plt.show()
```
